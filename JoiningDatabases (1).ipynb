```python
import pandas as pd
```

```python
inspection_letter_df = pd.read_excel('Joining Databases/inspection_letters.xlsx')
warning_letter_df = pd.read_csv("Joining Databases/warning_letter_final_data.csv")
other_letter_df = pd.read_excel('Joining Databases/483Num.xlsx')
```

```python
print(other_letter_df)
inspection_letter_df = inspection_letter_df.groupby('Legal Name',as_index=True).agg(lambda x: set(x)).rename_axis('Legal Name').reset_index()



```

```python
import string

def clean_string(s):
    # Remove punctuation
    s = str(s).translate(str.maketrans('', '', string.punctuation))
    # Convert to lowercase
    s = s.lower()
    
    # Remove Space
    s = s.replace(" ","")
    return s

```

```python
inspection_letter_df['clean name'] = inspection_letter_df['Legal Name'].apply(clean_string)
warning_letter_df['clean name'] = warning_letter_df['Company Name'].apply(clean_string)
other_letter_df['clean name'] =other_letter_df['Company Name'].apply(clean_string)
```

```python
merged_table = inspection_letter_df.merge(other_letter_df, on = 'clean name', how = 'outer')
merged_table = merged_table.merge(warning_letter_df, on = 'clean name', how = 'outer')

# merged_table = merged_table.merge(other_letter_df, on = 'clean name', how = 'inner')
merged_table = merged_table.drop('clean name', axis = 1)
merged_table
```
